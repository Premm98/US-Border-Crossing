```r
library(dplyr)
```

```r
countCanada <- function(data){
    count = 0
    for(x in data$Border){
        if(x == "US-Canada Border")
            count = count + 1
    }
    return(count)
}
```

```r
countMexico <- function(data){
    count1 = 0
    for(x in data$Border){
        if(x == "US-Mexico Border")
            count1 = count1 + 1
    }
    return(count1)
}
```

```r
totalCanadaVisits = countCanada(data)
print(totalCanadaVisits)
```

```r
totalMexicoVisits = countMexico(data)
print(totalMexicoVisits)
```

### Total Passengers that came to The US

```r
tot_pass <- pass_vehicle['Bus Passengers'] + pass_vehicle['Personal Vehicle Passengers'] + pass_vehicle['Train Passengers']
paste('Total Passengers came to The US:', tot_pass)
```

### Total passenger vehicle that entered The US

```r
tot_veh <- pass_vehicle['Buses'] + pass_vehicle['Personal Vehicles']
paste('Total Passenger vehicles that entered The US : ',tot_veh)
```

### Passenger to Vehicle ratio

```r
paste('Passenger to vehicle ratio: ',tot_pass/tot_veh)
```
